```python
#IMAGE THRESHOLDING >150 white,<150 black

import numpy as np
import cv2
from google.colab.patches import cv2_imshow



img=cv2.imread("/gdrive/My Drive/Colab Notebooks/img3.jpeg")
print(img.shape)
cv2_imshow(img)

gray1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(gray1.shape)
cv2_imshow(gray1)


ret, thresh1=cv2.threshold(gray1,130,255,cv2.THRESH_BINARY)
print(thresh1.shape)
cv2_imshow(thresh1)

ret, thresh1=cv2.threshold(gray1,130,255,cv2.THRESH_BINARY_INV)
print(thresh1.shape)
cv2_imshow(thresh1)

```

```python
#COLOR DETECTION IN AN IMAGE

##METHOD1(Extracting COlor channels and gray image)


#BLue Channel

import numpy as np
import cv2
from google.colab.patches import cv2_imshow

img=cv2.imread("/gdrive/My Drive/Colab Notebooks/img4.jpeg")
print("Imput Image")
cv2_imshow(img)

Blue_Channel1=img[:,:,0]
Green_Channel1=img[:,:,1]
Red_Channel1=img[:,:,2]

print("Blue Channel")
cv2_imshow(Blue_Channel1)

gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print("Gray Image")
cv2_imshow(gray)

Blue=cv2.subtract(Blue_Channel1,gray)
print("Subtracted Image")
cv2_imshow(Blue)


ret,thresh=cv2.threshold(Blue,110,255,cv2.THRESH_BINARY)
print("Threshold Image")
cv2_imshow(thresh)
```

```python
#Drawing Functions in OPENCV

import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow



#Create a black Image
img=np.zeros((512,512,3),np.uint8)#3D image taken

#Draw a Diagonal Blue Line thickness 5px

cv.line(img,(128,128),(256,256),(0,0,255),5)
cv2_imshow(img)

cv.rectangle(img,(384,0),(528,128),(255,0,0),5)
cv2_imshow(img)

cv.circle(img,(128,128),63,(255,0,0),1)
#Center of rectange nd radius, color, thickness
cv2_imshow(img)

```
